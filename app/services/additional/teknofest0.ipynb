```python
from difflib import get_close_matches
import unicodedata

# Function to remove accents and normalize non-Latin characters
def normalize_text(text):
    text = text.lower()
    text = unicodedata.normalize("NFKD", text)  # Normalize accented characters
    return "".join(c for c in text if not unicodedata.combining(c))  # Remove diacritics

def find_original_sentence(sentence, groups, threshold=0.5):
    lookup = {}
    all_phrases = []
    
    for group in groups:
        original = group[0]  # First element is the original
        for phrase in group:
            normalized = normalize_text(phrase)
            lookup[normalized] = original
            all_phrases.append(normalized)

    normalized_input = normalize_text(sentence)
    
    # Try exact match first
    if normalized_input in lookup:
        return lookup[normalized_input]
    
    # Try fuzzy matching
    close_matches = get_close_matches(normalized_input, all_phrases, n=1, cutoff=threshold)
    
    if close_matches:
        return lookup[close_matches[0]]  # Return original key
    
    return None  # No match found

# Predefined groups
groups = [
    ["akilli-ulasim-yarismasi", "smart-transportation-competition", "مسابقة المواصلات الذّكيّة", "Akıllı Ulaşım Yarışması", "Smart Transportation Competition"],
    ["biyoteknoloji-inovasyon-yarismasi", "biotechnology-innovation-competition", "مسابقة ابتكار التكنولوجيا الحيوية", "Biyoteknoloji İnovasyon Yarışması", "Biotechnology Innovation Competition"],
    ["blokzincir-yarismasi", "blockchain-competition", "مسابقة البلوكتشين", "Blokzincir Yarışması", "Blockchain Competition"],
    ["cevre-ve-enerji-teknolojileri-yarismasi", "environment-and-energy-technologies-competition", "مسابقة تقنيّات البيئة والطّاقة", "Çevre ve Enerji Teknolojileri Yarışması", "Environment and Energy Technologies Competition"],
]

# Test cases
sentences = [
    "cevre ve Enerji",  # Case difference
    "smart transportatioN",  # Case difference
    "çevre ve enerji teknolojileri",  # Turkish letters converted
    "BİYOTEKNOLOJİ inovasyon",  # Capital letters & partial
    "blok zincir",  # Space variation
    "ÇEVRE ve enerji",  # Mixed case
]

for s in sentences:
    print(f"{s} → {find_original_sentence(s, groups)}")
```
